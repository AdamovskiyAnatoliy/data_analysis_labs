```python
import numpy as np
```

```python
N = 10000
M = 1000
M_student = 1
MG = 51
trust_table = {
    50: 0.67,
    80: 1.28,
    90: 1.64,
    95: 1.96,
    99: 2.58
}
```

```python
y = np.random.uniform(10 + M_student, MG + 2 * M_student, N)
x = np.random.choice(y, M)
```

# Завдання А

```python
Y = y.sum()
Ys = y.mean()
DY2 = y.var(ddof=1)
print("""Сума елементів популяції: {}
Середнє значення популяції: {}
Десперсія популяції: {}""".format(Y, Ys, DY2))
```

# Завдання Б

```python
Y_ = x.sum() * N / M
Ys_ = x.mean()
S2 = x.var(ddof=1)
print("""Сума елементів популяції за вибіркою: {}
Середнє значення популяції за вибіркою: {}
Дисперсія популяції за вибіркою: {}""".format(Y_, Ys_, S2))
```

```python
DY_ = (S2 / M**2) * (1 - M / N ) * (N**2)
# DY_ = (S2 / M) * (1 - M / N ) * (N**2) 
DYs_ = (S2 / M)* (1 - M / N) 
print("""Дисперсія оцінок суми в залежності від об'єму вибірки: {}
Дисперсія оцінок середнього в залежності від об'єму вибірки: {}""".format(DY_, DYs_))
```

```python
Ys_interval = (Ys_ - trust_table[99] * np.sqrt(DYs_), Ys_ + trust_table[99] * np.sqrt(DYs_))
NYs_interval = (N * Ys_ - trust_table[99] * np.sqrt(DY_), N * Ys_ + trust_table[99] * np.sqrt(DY_))
print("""Довірчий інтервал для середнього: {}
Довірчий інтервал для суми: {}""".format(Ys_interval, NYs_interval))
```

# Завдання В

```python
N = 10000
M = [100, 1000]
```

```python
y = np.random.uniform(10 + M_student, MG + 2 * M_student, N)
```

```python
def generation_info(y, x, N, M):
    Y = y.sum()
    Ys = y.mean()
    DY2 = y.var(ddof=1)
    
    Y_ = x.sum() * N / M
    Ys_ = x.mean()
    S2 = x.var(ddof=1)
    
    DY_ = (S2 / M**2) * (1 - M / N ) * (N**2)
    # DY_ = (S2 / M) * (1 - M / N ) * (N**2) 
    DYs_ = (S2 / M)* (1 - M / N)
    
    Ys_interval = (Ys_ - trust_table[99] * np.sqrt(DYs_), Ys_ + trust_table[99] * np.sqrt(DYs_))
    NYs_interval = (N * Ys_ - trust_table[99] * np.sqrt(DY_), N * Ys_ + trust_table[99] * np.sqrt(DY_))
    return """
    ***
    Кількість елементів в вибірці: {}
    
    Сума елементів популяції: {}
    Середнє значення популяції: {}
    Дисперсія популяції: {}
    
    Сума елементів вибірки: {}
    Середнє значення вибірки: {}
    Дисперсія вибірки: {}
    
    Дисперсія оцінок суми в залежності від об'єму вибірки: {}
    Дисперсія оцінок середнього в залежності від об'єму вибірки: {}
    
    Довірчий інтервал для суми: {}
    Довірчий інтервал для середнього: {}
    """.format(M, Y, Ys, DY2, Y_, Ys_, S2, DY_, DYs_, NYs_interval, Ys_interval)
```

```python
for i in M:
    x = np.random.choice(y, i)
    print(generation_info(y, x, y.shape[0], i))
```

# Завдання Г

```python
for i in range(10, 2, -1):
    x = y[::i]
    print(generation_info(y, x, y.shape[0], x.shape[0]))
```

# Завдання Д

```python
def stratified_sampling(y, num_strat=10):
    random_choice = sorted(np.random.randint(0, len(y), num_strat-1))
    res = [y[:random_choice[0]]]
    for i in range(1, num_strat-1):
        res.append(y[random_choice[i-1]:random_choice[i]])
    res.append(y[random_choice[-1]:])
    return res
```

```python
num_strat = 10
y_strat = stratified_sampling(y, num_strat)
x_strat = []
for i in y_strat:
    x_strat.append(np.random.choice(i, int(len(i)/10)))
```

```python
w_h = []
for i in range(num_strat):
    w_h.append(len(y_strat[i])/len(x_strat[i]))
print('Вибіркова вага')
w_h
```

```python
y_mean_h = []
for i in y_strat:
    y_mean_h.append(i.mean())
print("Середнє значення для страти")
y_mean_h
```

```python
Y_h = []
for i in y_strat:
    Y_h.append(i.sum())
print('Загальна сума для страти')
Y_h
```

```python
Y = sum(Y_h)
print('Загальна сума для популяції')
Y
```

```python
y_u_ = Y/len(y)
print('середнє значення для всієї популяції')
y_u_
```

```python
S_2_h = []
for i in y_strat:
    S_2_h.append(i.var(ddof=1))
print('Дисперсія в страті')
S_2_h
```

```python
W_h = []
for i in y_strat:
    W_h.append(len(i)/len(y))
print('Вага страти')
W_h
```

```python
y__h_ = []
for i in x_strat:
    y__h_.append(i.mean())
print('Оцінка середнього для страти')
y__h_
```

```python
Y_h_ = []
for i in range(len(x_strat)):
    Y_h_.append(w_h[i]*x_strat[i].sum())
print('Оцінка загальної суми для страти')
Y_h_
```

```python
s_2_h = []
for i in x_strat:
    s_2_h.append(i.var(ddof=1))
print('Оцінка вибірквої дисперсії для страти')
s_2_h
```

```python
Y_st_ = sum(Y_h_)
print('Оцінка суми популяції')
Y_st_
```

```python
y_st__ = Y_st_ / len(y)
print('Оцінка середнього')
y_st__
```

```python
D_Y_st_ = []
for i in range(len(y_strat)):
    D_Y_st_.append((1 - 1/w_h[i])
                   * (len(y_strat[i])**2)
                   * (s_2_h[i]/len(x_strat[i])))
D_Y_st_ = sum(D_Y_st_)
print("Незміщення оцінка для дисперсії")
D_Y_st_
```

```python
D_y_st_ = []
for i in range(len(y_strat)):
    D_y_st_.append((1 - 1/w_h[i])
                   * ((len(y_strat[i])/len(y))**2)
                   * (s_2_h[i]/len(x_strat[i])))
D_y_st_ = sum(D_y_st_)
print("Незміщення оцінка для дисперсії")
D_y_st_
```

```python
print('Довірчий інтервал дял оцінки середнього значення')
(y_st__ - trust_table[99]*np.sqrt(D_y_st_), y_st__ + trust_table[99]*np.sqrt(D_y_st_))
```
